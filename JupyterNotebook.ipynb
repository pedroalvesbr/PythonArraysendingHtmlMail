```python
#Integração Python com Outlook
import win32com.client as win32
outlook= win32.Dispatch('outlook.application')
```

```python
#funçao pra ler os contatos do arquivo e retornar uma lista
def contatos(filename):
    names = []
    emails = []
    with open(filename, mode='r', encoding='utf-8') as contacts_file:
        for a_contact in contacts_file:
            names.append(a_contact.split()[0])
            emails.append(a_contact.split()[1])
    return names, emails 
```

```python
#apenas para testar o output
print(contatos('mycontacts.txt'))
```

```python
#return da funcao Contatos
names, emails = contatos('mycontacts.txt')

#corpo do email em html
email_html = open('mensagem.html')
email_body = email_html.read()
```

```python
#ENVIA EMAIL
for name, email in zip(names, emails):
    mail = outlook.CreateItem(0)
    mail.To= email
    mail.Subject ='ola ' +name+' esta é uma mensagem de teste'
    mail.HTMLBody = email_body
    mail.Send()
    print('email enviado para '+name)
```
